```python
from pathlib import Path
import pandas as pd
Path = "/content/drive/MyDrive/DATA_MINING_DATASETS/playlist_2010to2022.csv"

data = pd.read_csv(Path)

# Display the first few rows of the dataframe
data_head = data.head()

# Check data types and missing values
data_info = data.info()

# Summary statistics for the numerical columns
data_description = data.describe()

data_head, data_info, data_description





```

```python
# @ Let's proceed with these EDA steps, focusing first on missing value analysis and then on exploring categorical data.​
import matplotlib.pyplot as plt
import seaborn as sns

# Handling missing values
missing_values_count = data.isnull().sum()

# Categorical data analysis - focusing on unique value counts for some selected columns
unique_values = {
    column: data[column].nunique() for column in ['playlist_url', 'track_name', 'album', 'artist_name', 'artist_genres']
}

# Correlation analysis
correlation_matrix = data.select_dtypes(include=['float64', 'int64']).corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix of Numerical Features")
plt.show()

missing_values_count, unique_values

```

```python

# Temporal analysis - Looking at the trends of certain features over the years

# Selecting key features for temporal analysis
temporal_features = ['year', 'track_popularity', 'danceability', 'energy', 'loudness', 'tempo']

# Grouping by year and calculating mean for selected features
yearly_trends = data[temporal_features].groupby('year').mean()

# Plotting the trends
plt.figure(figsize=(15, 10))

for feature in temporal_features[1:]:
    sns.lineplot(data=yearly_trends, x='year', y=feature, label=feature)

plt.title("Yearly Trends of Audio Features and Track Popularity")
plt.ylabel("Mean Value")
plt.xlabel("Year")
plt.legend()
plt.grid(True)
plt.show()

# Distribution analysis - Looking at the distributions of key features
key_features = ['track_popularity', 'danceability', 'energy', 'loudness', 'tempo']

plt.figure(figsize=(15, 8))
for i, feature in enumerate(key_features, 1):
    plt.subplot(2, 3, i)
    sns.histplot(data[feature], bins=30, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()

yearly_trends.reset_index()


```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Step 1: Data Imputation
# Using median for imputation to avoid the influence of outliers
imputer = SimpleImputer(strategy='median')
data_imputed = imputer.fit_transform(data.select_dtypes(include=['float64', 'int64']))
data_imputed = pd.DataFrame(data_imputed, columns=data.select_dtypes(include=['float64', 'int64']).columns)

# Step 2: Normalization/Standardization
# Using StandardScaler for standardization
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_imputed)
data_scaled = pd.DataFrame(data_scaled, columns=data_imputed.columns)

data_scaled.head()

```

```python
from sklearn.cluster import KMeans

# Determining the optimal number of clusters using the Elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)

# Plotting the results of the Elbow method
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method For Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()

```

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Since the code execution state was reset, I'll reload the data and redo the preprocessing steps quickly

# Reload the dataset
Path = "/content/drive/MyDrive/DATA_MINING_DATASETS/playlist_2010to2022.csv"


# Load the dataset

data = pd.read_csv(Path)

# Impute missing values using the median
imputer = SimpleImputer(strategy='median')
data_imputed = imputer.fit_transform(data.select_dtypes(include=['float64', 'int64']))
data_imputed = pd.DataFrame(data_imputed, columns=data.select_dtypes(include=['float64', 'int64']).columns)

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_imputed)
data_scaled = pd.DataFrame(data_scaled, columns=data_imputed.columns)

# Clustering with K-means
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
cluster_labels = kmeans.fit_predict(data_scaled)

# Adding cluster labels to the scaled data
data_scaled['cluster'] = cluster_labels

# Anomaly Detection
# Calculating the distance of each point from its cluster centroid
centroids = kmeans.cluster_centers_
distance_from_centroid = [np.linalg.norm(x - centroids[cluster]) for x, cluster in zip(data_scaled.iloc[:, :-1].values, cluster_labels)]

# Adding the distance from centroid to the dataframe
data_scaled['distance_from_centroid'] = distance_from_centroid

# Sorting the data by distance from centroid to find potential anomalies
potential_anomalies = data_scaled.sort_values(by='distance_from_centroid', ascending=False)

# Displaying some of the potential anomalies
potential_anomalies.head()

```
